```python
#4
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('alphabet_stock_data.csv')

df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

start_date = '2004-09-10'
end_date = '2007-09-27'

filtered_data = df[start_date:end_date]

plt.figure(figsize=(12, 6))
plt.plot(filtered_data.index, filtered_data['Close'], label='Alphabet Inc. Stock Price')
plt.title('Alphabet Inc. Stock Price (Historical)')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('alphabet.csv')

df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

start_date = '2004-03-17'
end_date = '2016-03-17'

filtered_data = df[start_date:end_date]

plt.figure(figsize=(12, 6))
plt.bar(filtered_data.index, filtered_data['Volume'], label='Trading Volume')
plt.title('Alphabet Inc. Trading Volume (Historical)')
plt.xlabel('Date')
plt.ylabel('Trading Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('alphabet.csv')

df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

start_date = '2005-04-01'
end_date = '2005-05-01'

filtered_data = df[start_date:end_date]

plt.figure(figsize=(12, 6))
plt.scatter(filtered_data['Close'], filtered_data['Volume'], label='Trading Volume vs. Stock Price', alpha=0.5)
plt.title('Alphabet Inc. Stock Price vs. Trading Volume')
plt.xlabel('Stock Price (USD)')
plt.ylabel('Trading Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python

import pandas as pd

df = pd.read_csv('sales_data.csv')

pivot_table = df.pivot_table(index='Item', values='Sale_amt', aggfunc={'Sale_amt': ['max', 'min']})

pivot_table.columns = ['Max Sale', 'Min Sale']

print("Pivot Table - Maximum and Minimum Sale Values:")
print(pivot_table)
```

```python
import pandas as pd

df = pd.read_csv('sales_data.csv')

pivot_table = df.pivot_table(index='Item', values='Units', aggfunc='sum')

pivot_table.columns = ['Total Units Sold']

print("Pivot Table - Item-wise Total Units Sold:")
print(pivot_table)
```
